```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None) 
import warnings
warnings.simplefilter('ignore')
# para pintar
import pylab as plt
import seaborn as sns

# para que salga el grafico
%matplotlib inline  
```

```python
data_ori=pd.read_csv('../Data/attacks.csv',encoding="latin1")
data=data_ori.copy()
print(data.shape)
data
```

```python

plt.figure(figsize=(10, 6))

sns.heatmap(data.isna(),  # mapa de calor
            yticklabels=False,
            cmap='viridis',
            cbar=False);
```

```python
df = data.dropna(how='all', axis=0) 
```

```python
plt.figure(figsize=(10, 6))

sns.heatmap(df.isna(),  # mapa de calor
            yticklabels=False,
            cmap='viridis',
            cbar=False);
```
